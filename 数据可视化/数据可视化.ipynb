```python
import networkx as nx
from pylab import *
from networkx.algorithms import approximation as approx
mpl.rcParams['font.sans-serif'] = ['SimHei'] 
# import matplotlib.pyplot as plt

G = nx.DiGraph()
# G.add_node('中国')     # 添加节点z
G.add_nodes_from([1, 2, 3,4,5])   # 添加节点 1 2 3
# G.add_edge('x', 'y', weight=15, label='123')          # 添加边  起点为x  终点为y
G.add_edges_from([(1, 2),(2,3), (4,3),(1,4) ],label="WN2898")   # 添加多条边

# 最短路径
# print(nx.dijkstra_path(G, 1, 4))
print(G.has_edge(1,2))
print(G.edges())
# 寻找节点之间连接数
print(approx.node_connectivity(G,4,1))

# 到节点3的所有相关节点
print(list(nx.all_simple_paths(G,1,3)))

nx.draw_circular(G, with_labels=True)
plt.show()
# 网络图绘制与显示
# list(nx.k_edge_components(G, k=4))

# sorted(map(sorted, nx.k_edge_components(G, k=1)))
# from networkx.algorithms.boundary import node_boundary
# node_boundary(G,[1,3])
```
